```python
import pygame
from tkinter import *
from tkinter import filedialog
from pytube import YouTube

# Initialize pygame
pygame.init()

# Create the main window
root = Tk()
root.title("Music Player")

# Global variables
playlist = []
current_song = 0

# Function to open files and create playlist
def open_files():
    file_paths = filedialog.askopenfilenames(filetypes=[("Audio Files", "*.mp3 *.wav")])
    playlist.extend(file_paths)
    if not pygame.mixer.music.get_busy():
        play_music(current_song)

# Function to play music
def play_music(song_index):
    global current_song
    if song_index < 0:
        current_song = len(playlist) - 1
    elif song_index >= len(playlist):
        current_song = 0
    file_path = playlist[current_song]
    pygame.mixer.music.load(file_path)
    pygame.mixer.music.play()
def display_playlist():
    playlist_window = Toplevel(root)
    playlist_window.title("Playlist")
    
    playlist_text = Text(playlist_window)
    for index, song in enumerate(playlist, start=1):
        playlist_text.insert(END, f"{index}. {song}\n")
    playlist_text.pack()
# Function to pause music
def pause_music():
    pygame.mixer.music.pause()

# Function to unpause music
def unpause_music():
    pygame.mixer.music.unpause()

# Function to stop music
def stop_music():
    pygame.mixer.music.stop()

# Function to skip to the next song
def next_song():
    global current_song
    current_song += 1
    play_music(current_song)

# Function to skip to the previous song
def prev_song():
    global current_song
    current_song -= 1
    play_music(current_song)

# Function to download audio from YouTube link
def download_youtube_audio():
    try:
        yt = YouTube(video_url)
        video_stream = yt.streams.get_highest_resolution()
        video_stream.download("downloads")
        
        # Get the downloaded video file name
        video_filename = "downloads/" + video_stream.default_filename
        
        # Convert to MP3
        mp3_filename = video_filename.replace(".mp4", ".mp3")
        video_clip = VideoFileClip(video_filename)
        audio_clip = video_clip.audio
        audio_clip.write_audiofile(mp3_filename)
        audio_clip.close()
        video_clip.close()
        
        # Delete the original video (MP4) file
        os.remove(video_filename)
        
        playlist.append(mp3_filename)
        if not pygame.mixer.music.get_busy():
            play_music(current_song)
        status_label.config(text="Downloaded and added to playlist.")
    except Exception as e:
        status_label.config(text="Error downloading: " + str(e))
# Create UI elements
open_button = Button(root, text="Open", command=open_files)
open_button.pack()

play_button = Button(root, text="Play", command=lambda: play_music(current_song))
play_button.pack()

pause_button = Button(root, text="Pause", command=pause_music)
pause_button.pack()

unpause_button = Button(root, text="Unpause", command=unpause_music)
unpause_button.pack()

stop_button = Button(root, text="Stop", command=stop_music)
stop_button.pack()

prev_button = Button(root, text="Previous", command=prev_song)
prev_button.pack()

next_button = Button(root, text="Next", command=next_song)
next_button.pack()

display_playlist_button = Button(root, text="Display Playlist", command=display_playlist)
display_playlist_button.pack()

youtube_label = Label(root, text="Enter YouTube URL:")
youtube_label.pack()

youtube_entry = Entry(root, width=50)
youtube_entry.pack()

download_button = Button(root, text="Download from YouTube", command=download_youtube_audio)
download_button.pack()

status_label = Label(root, text="")
status_label.pack()

# Run the GUI main loop
root.mainloop()

# Quit pygame
pygame.quit()
```
